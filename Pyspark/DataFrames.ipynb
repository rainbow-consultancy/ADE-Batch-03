```python
# 1. DataFrame withOut strict schema

my_data = [
    (100, 'Sandeep', 76),
    (101, 'Nikhil', 86),
    (102, 'Gowri', 90),
    (103, 'Ravi', 80)
]

columns = ["stu_id", "stu_name", "marks"]

stu_df = spark.createDataFrame(my_data, columns)
display(stu_df)
```

```python
# 2. Create DataFrame with strict schema
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

emp_data = [
    (100, "Lokesh", "IT", 30000),
    (101, "Ravi", "HR", 40000),
    (102, "Rajesh", "IT", 35000),
    (103, "Ramesh", "HR", 45000)
]

schema = StructType([
    StructField("emp_id", IntegerType(), False),
    StructField("name", StringType(), True),
    StructField("dept", StringType(), True),
    StructField("salary", IntegerType(), True)
])

emp_df = spark.createDataFrame(emp_data, schema)
display(emp_df)
```

```python
df = (
    spark.read.format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load("/FileStore/tables/Students_Social_Media_Addiction.csv")
)

display(df)
```

```python
# where clause - filters the data

high_sch_df = df.filter(df.Academic_Level == "High School") 
high_sch_df.display()
```

```python
high_sch_df = df.where(df.Academic_Level == "High School") 
high_sch_df.display()
```
